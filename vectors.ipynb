# Vector-scalar multiplication

```python
# vector and scalar
v1 = np.array([3, -1])
l = -0.3
v1m = v1 * l # scalar modulated

# plot them



plt.plot([0, v1[0]], [0, v1[1]], label='Original vector')
plt.plot([0, v1m[0]], [0, v1m[1]], 'r:', label='Scaled vector')

axlim = max([abs(max(v1)), abs(max(v1m))]) * 1.5  # dynamic axis lim
plt.axis('square')
plt.axis((-axlim, axlim, -axlim, axlim))
plt.plot([0, 0], [-axlim, axlim], 'k--')
plt.plot([-axlim, axlim], [0, 0], 'k--')
plt.grid()
plt.legend()
plt.show()
```

# Vector-vector multiplication: the dot product

```python
## many ways to compute the dot product

v1 = np.array([1, 2, 3, 4, 5, 6])
v2 = np.array([0, -4, -3, 6, 5, 1])

# method 1
dp1 = sum(np.multiply(v1, v2)) # sum element-wise multipliation of two vectors

# method 2
dp2 = np.dot(v1, v2)

# method 3 - matrix multiplication
dp3 = np.matmul(v1, v2)

# method 4 - looping over elements
dp4 = 0 # initialize

#== loop over elements
for i in range(0, len(v1)):
    
    # multiply corresponding element and sum
    dp4 += v1[i] * v2[i]
    
print(dp1, dp2, dp3, dp4)
```

# Dot product properties: associative and distributive

```python
# Distributive property

# create random vectors
n = 5
# generate a "n" element random vector. Each element is drawn from a normal distribution of random numbers.
v1 = np.random.randn(n)
v2 = np.random.randn(n)
v3 = np.random.randn(n)

# the two results
d_res1 = np.dot(v1, v2+v3)
d_res2 = np.dot(v1, v2) + np.dot(v1, v3)

print("Distributive property:")
print([d_res1, d_res2])

# Associative property

a_res1 = np.dot(np.dot(v1, v2), v3) # technically this is a vector-scalar multiplication and not a dot product
a_res2 = np.dot(v1, np.dot(v2, v3))

print("Associative property:")
print(a_res1)
print(a_res2)
```

# Code challenge: dot products with matrix columns

```python
# coding challenge

m1 = np.random.randn(4, 6)
m2 = np.random.randn(4, 6)

# Method 1
dots = []

for j in range(6): # col
    res = 0
    for i in range(4): # row
        res += m1[i][j] * m2[i][j]
    dots.append(res)
    
print(dots)

# Method 2

dots1 = np.zeros(6)

for i in range(6):
    dots1[i] = np.dot(m1[:,i], m2[:,i])

print(' ')
print(dots1)
    
```

# Dot product properties: commutative

```python
n = 5
# generate a "n" element random vector. Each element is drawn from a normal distribution of random numbers.
v1 = np.random.randn(n)
v2 = np.random.randn(n)

# Commutative property

c_res1 = np.dot(v1, v2)
c_res2 = np.dot(v2, v1)

print("Commutative property:")
print([c_res1, c_res2])
```

# Vector length

```python

v1 = np.array([1, 2])

# methods 1-4, just like with regular dot product
vl1 = np.sqrt(np.dot(v1, v1))
print(vl1)

# method 5: take the norm
vl2 = np.linalg.norm(v1)
print(vl2)
```
